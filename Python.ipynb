Pertanyaan No 1

```python
class Order:  # membuat class bernama Order sebagai template untuk data pesanan

    # 1. Fungsi inisialisasi
    def __init__(self, order_id, customer_name, order_date, total_amount):  # fungsi yang otomatis dijalankan saat objek dibuat
        self.order_id = order_id        # menyimpan nilai order_id ke dalam objek
        self.customer_name = customer_name  # menyimpan nama pelanggan ke dalam objek
        self.order_date = order_date    # menyimpan tanggal pesanan
        self.total_amount = total_amount  # menyimpan total harga pesanan

    # 2. Fungsi hitung pajak
    def calculate_tax(self, tax_rate):  # fungsi untuk menghitung pajak berdasarkan tarif pajak
        return self.total_amount * tax_rate  # mengalikan total_amount dengan tax_rate lalu mengembalikan hasilnya

    # 3. Fungsi tampil
    def display_order(self):  # fungsi untuk menampilkan informasi pesanan
        print(f"ID: {self.order_id} | Nama: {self.customer_name} | Total: {self.total_amount}")
        # menampilkan ID pesanan, nama pelanggan, dan total harga menggunakan f-string
```

```python
order1 = Order(101, "Rachel", "2026-03-30", 500000)
# membuat objek pesanan bernama order1 dari class Order
# 101 = ID pesanan
# "Rachel" = nama pelanggan
# "2026-03-30" = tanggal pesanan
# 500000 = total harga pesanan

order1.display_order()
# memanggil fungsi display_order() dari objek order1
# fungsi ini akan menampilkan data pesanan (ID, nama, total)

tax = order1.calculate_tax(0.1)
# memanggil fungsi calculate_tax() untuk menghitung pajak
# 0.1 artinya pajak 10%
# hasil perhitungan pajak disimpan ke variabel tax

print("Pajak:", tax)
# menampilkan tulisan "Pajak:" dan nilai pajak yang sudah dihitung
```

```python
class OrderProcessor:  # membuat class bernama OrderProcessor untuk mengelola banyak pesanan

    # 1. Tempat penyimpanan
    def __init__(self):  # fungsi yang otomatis dijalankan saat objek OrderProcessor dibuat
        self.orders = []  # membuat list kosong bernama orders untuk menyimpan semua objek Order

    # 2. Fungsi menambah pesanan
    def add_order(self, order):  # fungsi menerima parameter order (objek Order)
        self.orders.append(order)  # menambahkan objek order ke dalam list orders

    # 3. Fungsi hitung total pendapatan
    def calculate_total_revenue(self):  # fungsi untuk menghitung total pendapatan dari semua pesanan
        total = 0  # membuat variabel total untuk menyimpan jumlah pendapatan

        for order in self.orders:  # melakukan perulangan untuk setiap objek order di dalam list orders
            total += order.total_amount  # menambahkan nilai total_amount dari setiap order ke variabel total

        return total  # mengembalikan hasil total pendapatan

    # 4. Fungsi hitung total pajak
    def calculate_total_tax(self, tax_rate):  # fungsi untuk menghitung total pajak dari semua pesanan
        total_tax = 0  # membuat variabel untuk menyimpan total pajak

        for order in self.orders:  # melakukan perulangan untuk setiap pesanan
            total_tax += order.calculate_tax(tax_rate)  # memanggil fungsi calculate_tax dari setiap order lalu menjumlahkannya

        return total_tax  # mengembalikan hasil total pajak
```

```python
order1 = Order(1, "Rachel", "2026-03-30", 500000)
# membuat objek pesanan pertama dari class Order
# 1 = ID pesanan
# "Rachel" = nama pelanggan
# "2026-03-30" = tanggal pesanan
# 500000 = total harga pesanan

order2 = Order(2, "Budi", "2026-03-30", 300000)
# membuat objek pesanan kedua
# 2 = ID pesanan
# "Budi" = nama pelanggan
# tanggal sama
# total pesanan 300000

processor = OrderProcessor()
# membuat objek dari class OrderProcessor
# objek ini berfungsi untuk mengelola dan menyimpan banyak pesanan

processor.add_order(order1)
# menambahkan pesanan order1 ke dalam list orders di OrderProcessor

processor.add_order(order2)
# menambahkan pesanan order2 ke dalam list orders di OrderProcessor

print("Total Revenue:", processor.calculate_total_revenue())
# memanggil fungsi calculate_total_revenue()
# fungsi ini menjumlahkan total_amount dari semua pesanan
# 500000 + 300000 = 800000

print("Total Tax:", processor.calculate_total_tax(0.1))
# memanggil fungsi calculate_total_tax() untuk menghitung total pajak
# 0.1 = pajak 10%
# pajak order1 = 500000 × 0.1 = 50000
# pajak order2 = 300000 × 0.1 = 30000
# total pajak = 80000
```
